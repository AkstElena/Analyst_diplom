```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Загружаем сохраненный после парсинга сайта Ozon, файл json.
Выводим его заголовок

```python
df = pd.read_json('ozon_one_page_list_all_page_10-10-2024.json')
df.head()
```

Удаляем полностью дублирующие друг друга строки

```python
df = df.drop_duplicates()
df
```

Изучаем данные в таблице: какие строки, количество данных, тип данных

```python
df.info()
```

Анализируем количество пустых ячеек. На основе этого анализа нужно принять решение какие строки оставить, а какие убрать

```python
round(df.isnull().sum()*100/len(df), 2).sort_values(ascending=False)
```

Видим что большая значительная часть столбцов есть только у малого процента позиций, поэтому данные столбцы следует исключить из данных. Они не дадут полезной информации, так как не дадут сравнить между собой позиции. Таким образом удаляем практически все столбцы, где нулевых значений больше 80%, за исключение столбца "Общий объем HDD, ГБ", так как это необходимая информация.

Так как столбцов больше на удаление, логичнее создать новый датафрейм только с необходимыми столбцами

```python
new_df = df[['Наименование', 'Цена по карте Озон', 'Оперативная память', 'Общий объем SSD, ГБ', 'Общий объем HDD, ГБ', 'Операционная система', 'Процессор', 'Видеокарта', 'Дата']]
new_df.head()
```

Снова смотрим информацию по полученным данным

```python
new_df.info()
```

Видим, что позиция оперативная память не числовая, а строковая, стоит это исправить. Для этого изучаем какие есть уникальные строки в столбце оперативная память.

```python
new_df['Оперативная память'].value_counts()
```

Так как все строки включат размер в ГБ, есть смысл перенести ГБ в название столбца и оставить только цифровые значения

```python
new_df['Оперативная память'] = new_df['Оперативная память'].str.split(' ').str.get(0)
new_df
```

```python
new_df.rename(columns={"Оперативная память": "Оперативная память, ГБ"}, inplace=True)
```

Преобразуем текстовый вид данных в цифровые

```python
new_df['Оперативная память, ГБ'] = new_df['Оперативная память, ГБ'].astype('float')
```

Далее снова изучаем данные

```python
new_df.info()
```

Далее изучаем в каких данных нет информации по оперативной памяти

```python
new_df[new_df['Оперативная память, ГБ'].isnull()]
```

Видим что кроме ноутбуков также встречаются наклейки, подставки и чехлы. Их необходимо исключить.

```python
new_df[new_df[['Оперативная память, ГБ', 'Общий объем SSD, ГБ', 'Общий объем HDD, ГБ', 'Процессор', 'Видеокарта', 'Операционная система']].isna().all(axis=1)]
```

Удаляем ячейки, где все нижеперечисленные ячейки пустые

```python
new_df = new_df.dropna(subset=['Оперативная память, ГБ', 'Операционная система', 'Процессор', 'Видеокарта', 'Общий объем SSD, ГБ', 'Общий объем HDD, ГБ'], how='all')
```

```python
new_df.info()
```

Смотрим статистические данные по полученной информации, приводя среднее значение к двум знакам после запятой

```python
new_df.describe().applymap(lambda x: f'{x:0.2f}')
```

По цене видны выбросы, следует проанализировать их графически

```python
sns.boxplot(new_df['Цена по карте Озон'])
plt.ylabel('Цена, руб.')
plt.title('Разброс цены ноутбуков');
```

Видим графически достаточной большой выброс по дорогим ценам. Для их исключения уберем 2,5% с обоих сторон

```python
new_df = new_df[(new_df['Цена по карте Озон'] < new_df['Цена по карте Озон'].quantile(0.975)) & (new_df['Цена по карте Озон'] > new_df['Цена по карте Озон'].quantile(0.025))]
```

```python
sns.boxplot(new_df['Цена по карте Озон'])
plt.ylabel('Цена, руб.')
plt.title('Разброс цены ноутбуков');
```

Снова смотрим статистические значения

```python
new_df.describe().applymap(lambda x: f'{x:0.2f}')
```

Анализируем столбцы, связанные с размером памяти. Находим те, где отсутствуют данные и по SSD и по HDD

```python
new_df[(new_df['Общий объем SSD, ГБ'].isnull()) & (new_df['Общий объем HDD, ГБ'].isnull())]
```

Меняем данные по факту (в дальнейшем продумаю функцию обработки данных из наименования)

```python
new_df.loc[22, 'Общий объем SSD, ГБ'] = 256.0
new_df.loc[95, 'Общий объем SSD, ГБ'] = 512.0
new_df.loc[242, 'Общий объем SSD, ГБ'] = 512.0
new_df.loc[242, 'Оперативная память, ГБ'] = 16.0
new_df.loc[284, 'Общий объем SSD, ГБ'] = 512.0
new_df.loc[323, 'Общий объем SSD, ГБ'] = 512.0
```

Убеждаемся что больше нет одновременно пустых ячеек по данным памяти

```python
new_df[(new_df['Общий объем SSD, ГБ'].isnull()) & (new_df['Общий объем HDD, ГБ'].isnull())]
```

Теперь заменяем все отсутствующие значения в столбцах Общий объем на 0

```python
new_df['Общий объем SSD, ГБ'] = new_df['Общий объем SSD, ГБ'].fillna(0)
new_df['Общий объем HDD, ГБ'] = new_df['Общий объем HDD, ГБ'].fillna(0)
```

```python
round(new_df.isnull().sum()*100/len(new_df), 2).sort_values(ascending=False)
```

Так же есть отсутствующие данные по оперативной памяти

```python
new_df[new_df['Оперативная память, ГБ'].isnull()]
```

Заменим эти данные на наиболее часто встречающиеся

```python
new_df['Оперативная память, ГБ'].fillna(new_df['Оперативная память, ГБ'].median(), inplace=True)
```

```python
new_df[new_df['Оперативная память, ГБ'].isnull()]
```

```python
round(new_df.isnull().sum()*100/len(new_df), 2).sort_values(ascending=False)
```

По остальным пустым позициям поставим текстовое значение: Нет данных

```python
new_df.fillna(value = 'Нет данных', inplace = True)
```

```python
round(new_df.isnull().sum()*100/len(new_df), 2).sort_values(ascending=False)
```

Таких находится в данной базе 5 позиций, при этом в самом описании товара, размеры памяти указаны

```python
plt.hist(new_df['Цена по карте Озон'])
plt.title('Стоимость ноутбуков')
plt.xlabel('Цена ноутбука')
plt.ylabel('Количество объектов');
```

```python
sns.jointplot(x=new_df['Цена по карте Озон'], y=new_df['Общий объем SSD, ГБ'], kind='reg')

plt.xlabel('Цена')
plt.ylabel('Оперативная память')
plt.title('Зависимость цены от общего объема памяти');
```
